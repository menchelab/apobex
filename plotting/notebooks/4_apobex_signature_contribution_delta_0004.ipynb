```python
def filter_and_process_data(merged_data, clone, days_in_culture):
    # Filter the merged data based on the clone and days_in_culture
    filtered_data = merged_data[(merged_data['clone'] == clone) & (merged_data['days_in_culture'] == days_in_culture)]
    
    genotype_order = ['sgAAVS1', 'sgUBR4', 'sgUBR5', 'sgHUWE1']
    # Convert the 'E3_genotype' column to a categorical type with the specified order
    filtered_data['E3_genotype'] = pd.Categorical(filtered_data['E3_genotype'], categories=genotype_order, ordered=True)
    
    # Sort the dataframe by the 'E3_genotype' column
    sorted_data = filtered_data.sort_values('E3_genotype')
    
    # Select the relevant columns
    processed_data = sorted_data[['sample_ID', 'SBS1', 'SBS5', 'SBS6', 'SBS15', 'SBS18', 'SBS13', 'SBS2']]
    
    return processed_data
```

```python
#filter the merged data so i only have day 6 and day 10 for each clone
RA1 = filter_and_process_data(merged_data, 'RA1', 'D610')
R2 = filter_and_process_data(merged_data, 'R2', 'D610')
R2_219 = filter_and_process_data(merged_data, 'R2+219', 'D610')

df_list = [RA1, R2, R2_219]
```

```python
def average_replicates(input_dataframe):
    input_dataframe.reset_index(drop=True, inplace=True)
    
    # Convert relevant columns to numeric
    numeric_columns = input_dataframe.columns.drop('sample_ID')
    input_dataframe[numeric_columns] = input_dataframe[numeric_columns].apply(pd.to_numeric, errors='coerce')
    
    # Calculate the average for all columns, 3 rows at a time, excluding 'sample_ID'
    averaged_data = input_dataframe[numeric_columns].groupby(input_dataframe.index // 3).mean()
    
    # Add a new column 'sample' with the specified values
    sample_values = ['control', 'UBR4', 'UBR5', 'HUWE1']
    averaged_data['sample'] = sample_values
    
    return averaged_data
```

```python
# make average replicates for each clone and day
RA1_avg = average_replicates(filter_and_process_data(merged_data, 'RA1', 'D610'))
R2_avg = average_replicates(filter_and_process_data(merged_data, 'R2', 'D610'))
R2_219_avg = average_replicates(filter_and_process_data(merged_data, 'R2+219', 'D610'))


#make a list of the average replicates
avg_df_list = [RA1_avg, R2_avg, R2_219_avg]

# Normalize each dataframe in avg_df_list by row sums
normalized_avg_df_list = []
for df in avg_df_list:
    row_sums = df.iloc[:, :-1].sum(axis=1)
    normalized_df = df.iloc[:, :-1].div(row_sums, axis=0)
    normalized_df['sample'] = df['sample']
    normalized_avg_df_list.append(normalized_df)
```

```python
# plot the averaged replicates

for df, df_name in zip(normalized_avg_df_list, df_names):
    pu.plot_genotype_comparison_summary(df, plot_title=f'Signature Activity {df_name}')

```

```python
RA1_stats = avg_df_list[0]
R2_stats = avg_df_list[1]
R2_219_stats = avg_df_list[2]

stats_df_list = [RA1_stats, R2_stats, R2_219_stats]

for df, df_name in zip(stats_df_list, df_names):

    #round anything to whole numbers except the sample column
    df.iloc[:, :-1] = df.iloc[:, :-1].round(0)

    #make a new column that calculates the total mutations for each sample (row sums)
    df['total_mutations'] = df.iloc[:, :-1].sum(axis=1)

    #make a new colum that calculates the sum of SBS2 and SBS13 mutations (that is column index 5 and 6, per row)
    df['APOBEC_signature'] = df.iloc[:, 5:7].sum(axis=1)

    #make a new column that calculates the ratio of APOBEC mutations to total mutations
    df['APOBEC_ratio'] = df['APOBEC_signature'] / df['total_mutations']

    #make a new column that expressed the apobex ratio as percent
    df['APOBEC_ratio_percent'] = df['APOBEC_ratio'] * 100


stats_df_list[0]
```

```python
# Extract the control data from each dataframe
controls = [df[df['sample'] == 'control'] for df in stats_df_list]

# Extract the APOBEC_ratio_percent for each control
apobec_ratios = [control['APOBEC_ratio_percent'].values[0] for control in controls]


# Plot the bar plot
plt.figure(figsize=(2.5, 4))
plt.bar(df_names, apobec_ratios, color='lightgreen')
plt.xlabel('Clone')
plt.ylabel('Percent APOBEC mutations')
plt.title('APOBEC Ratio Percent for Controls')
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import fisher_exact

# Function to perform Fisher's exact test
def perform_fishers_exact_test(df, control_label='control'):
    results = []
    control_data = df[df['sample'] == control_label]
    
    for sample in df['sample'].unique():
        if sample == control_label:
            continue
        
        sample_data = df[df['sample'] == sample]
        
        # Construct the contingency table
        contingency_table = [
            [control_data['APOBEC_signature'].sum(), control_data['total_mutations'].sum() - control_data['APOBEC_signature'].sum()],
            [sample_data['APOBEC_signature'].sum(), sample_data['total_mutations'].sum() - sample_data['APOBEC_signature'].sum()]
        ]

        print(contingency_table)
        
        # Perform Fisher's exact test
        odds_ratio, p_value = fisher_exact(contingency_table)
        
        results.append({
            'sample': sample,
            'odds_ratio': odds_ratio,
            'p_value': p_value
        })
    
    return pd.DataFrame(results)

# Apply the function to each dataframe in stats_df_list
fishers_results_list = []
for df, df_name in zip(stats_df_list, df_names):
    fishers_results = perform_fishers_exact_test(df)
    fishers_results['clone'] = df_name
    fishers_results_list.append(fishers_results)

# Combine all results into a single dataframe
combined_fishers_results = pd.concat(fishers_results_list, ignore_index=True)

combined_fishers_results['bonferroni_corrected_p_value'] = combined_fishers_results['p_value'] * 3
combined_fishers_results['bonferroni_corrected_p_value'] = combined_fishers_results['bonferroni_corrected_p_value'].clip(upper=1.0)
combined_fishers_results['significant'] = (combined_fishers_results['bonferroni_corrected_p_value'] < 0.05) & (combined_fishers_results['odds_ratio'] < 1)

combined_fishers_results

combined_fishers_results.to_csv('/Users/mathilde/Desktop/projects/APOBEX/4_apobex/APOBEX_4.4/plots/fishers_results.csv', sep=',', index=False)
```
